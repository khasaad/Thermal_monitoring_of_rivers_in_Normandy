# Suivi thermique des rivères en Normandie

La Touques, l’Orne et la Sélune ont été équipées par la DREAL de Normandie d’un suivi thermique en juin 2011. Ce suivi a été mis en oeuvre pour approfondir la connaissance globale des écosystèmes fluviatiles de Normandie et pour comprendre l’impact des nappes d’eau souterraines et des barrages de la Suisse normande sur les régime thermiques des deux rivières instrumentées.

Les sondes sont attachées aux racines des arbres de la ripisylve, en position ombragée. Elles sont installées à une profondeur garantissant leur immersion complète quelle que soit la saison. La température est mesurée toutes les deux heures. Les sondes utilisées sont des HOBO® water Temp Pro v2 (U22-001). Leur précision est de +/-0,2°C. Les données peuvent être lues grâce à deux interfaces : optique et USB.

## Pourquoi un suivi thermique

La température des cours d'eau joue un rôle fondamental dans la dynamique des écosystèmes aquatiques : elle influe sur la distribution spatiale des organismes vivants, leur croissance, leur reproduction, leur comportement, dont la migration ou encore leur état sanitaire. Cette description physique est donc une première étape de la connaissance biologique des cours d’eau.  

En effet, les espèces aquatiques choisissent leurs habitats en tenant compte de combinaisons de facteurs. Parmi eux, principalement, la disponibilité physiologique de l’oxygène, qui dépend directement de la température et du mouvement de l’eau.

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np

name = 'data_monne.csv'

df6 = pd.read_csv(name)

def plot_temperature_ville(df, year, second_year):

    axis_x = []
    axis_y_mean = []
    axis_y_min = []
    axis_y_max = []
    
    
    if year is not None and second_year is None: 
        for n1 in range(len(df)):
            if df.loc[n1, 'date'].startswith(year):
                axis_x.append(df.loc[n1, 'date'])
                axis_y_mean.append(df.loc[n1, 'Teau_mean'])
                axis_y_min.append(df.loc[n1, 'Teau_min'])
                axis_y_max.append(df.loc[n1, 'Teau_max'])
                
    elif year is None and second_year is None:
        for n2 in range(len(df)):
            axis_x.append(df.loc[n2, 'date'])
            axis_y_mean.append(df.loc[n2, 'Teau_mean'])
            axis_y_min.append(df.loc[n2, 'Teau_min'])
            axis_y_max.append(df.loc[n2, 'Teau_max'])
            
    elif year is not None and second_year is not None:
        for n1 in range(len(df)):
            if df.loc[n1, 'date'].startswith(year) or df.loc[n1, 'date'].startswith(second_year):
                axis_x.append(df.loc[n1, 'date'])
                axis_y_mean.append(df.loc[n1, 'Teau_mean'])
                axis_y_min.append(df.loc[n1, 'Teau_min'])
                axis_y_max.append(df.loc[n1, 'Teau_max'])
            
    else:
        print('second_year argument must not be equal an year if year argument is None')

#     print(len(axis_x), '*', len(axis_y_mean), '*', len(axis_y_min), '*', len(axis_y_max))

    fig = plt.figure(figsize=(20, 10))

    ax = fig.add_subplot(1, 1, 1)

    plt.xlabel('date')
    plt.title('Représentation graphique des températures moyennes (T°moy), minimales (T°min), maximales (T°max)\
    à la journée sur la satation ' + name[5: name.index('.')] + ' – données nettoyées.')

    plt.ylabel('temperature °c')
    ax.plot(axis_x, axis_y_min, label='min', color='aqua')
    ax.plot(axis_x, axis_y_max, label='max', color='fuchsia')
    ax.plot(axis_x, axis_y_mean, label='mean', color='orange')
    ax.legend(loc='best')
    plt.hlines(20, 0, len(axis_x), colors='red', label='temp = 20 °c')
    plt.hlines(4, 0, len(axis_x), colors='blue', label='temp = 4 °c')
    plt.xticks([p for p in range(0, len(axis_x), 50)], rotation=60, color='green', size="x-large")
    plt.legend(loc='best')
```

```python
import pandas as pd 


df6 = pd.read_csv('data_monne.csv')

plot_temperature_ville(df6, year='2013', second_year=None)
plot_temperature_ville(df6, year='2015', second_year='2016')
plot_temperature_ville(df6, year=None, second_year=None)
```

## Afficher bar 3D qui représente les nombres des jours où la température inférieur à 4 °c

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm


# Extraire data avec température inférieur à 4 °c

result = []

for k in Dict_bar.keys():
    list_val = []
    for v in Dict_bar[k].keys():
        list_val.append(Dict_bar[k][v]['TempInf4'])

    result.append(list_val)
    

result = np.array(result, dtype=np.int)


fig=plt.figure(figsize=(5, 5), dpi=150)
ax1=fig.add_subplot(111, projection='3d')

xlabels = np.array(['2013', '2014', '2015',
                    '2016', '2017', '2018'])
xpos = np.arange(xlabels.shape[0])
ylabels = np.array(['Monne', 'Barge', 'Ving Bec', 'Odon T2', 'Odon T4', 'Odon T5', 'Orne T1',
                    'Orne T3', 'Orne T2', 'Selune T4', 'Selune T2', 'Selune T5', 'Selune T1', 
                    'Touques T1', 'Touques T3', 'Touques T4', 'Touques T6'])
ypos = np.arange(ylabels.shape[0])

xposM, yposM = np.meshgrid(xpos, ypos, copy=False)

zpos=result
zpos = zpos.ravel()

dx=0.5
dy=0.5
dz=zpos

ax1.w_xaxis.set_ticks(xpos + dx/2.)
ax1.w_xaxis.set_ticklabels(xlabels, rotation=60)

ax1.w_yaxis.set_ticks(ypos + dy/2.)
ax1.w_yaxis.set_ticklabels(ylabels, rotation=330, fontsize = 5, va='bottom', ha='left')

values = np.linspace(0.2, 1., xposM.ravel().shape[0])
colors = cm.rainbow(values)
ax1.bar3d(xposM.ravel(), yposM.ravel(), dz*0, dx, dy, dz, color=colors)
plt.title('Nombres des jours où la température inférieur à 4 °c')
plt.show()
```
